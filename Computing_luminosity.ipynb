```python
# Computing the distance of the source in relation to Earth

#Paramenters
parallax_mas = 32.0589          # Gaia DR3
parallax_err = 0.1458


#Conversion to Parcec
distance_pc = 1000 / parallax_mas

#Defining the Funtion for distance
distance_err_pc = (
    parallax_err / parallax_mas
) * distance_pc

#Print distance 
print(f"Distance = {distance_pc:.3f} ± {distance_err_pc:.3f} pc")
```

```python
#Computing the Luminosities from the fluxes

# --------------------------------------------------
# Convert pc -> cm
# --------------------------------------------------
pc_to_cm = 3.085677581e18
distance_cm = distance_pc * pc_to_cm

# --------------------------------------------------
# ROSAT fluxes (0.1-2.4 keV)
# --------------------------------------------------

fluxes = {"Flux1": 2.06e-12,   # erg cm^-2 s^-1
          "Flux2": 1.70e-12}

# --------------------------------------------------
# Luminosities
# --------------------------------------------------

for name, Fx in fluxes.items():

    Lx = 4 * np.pi * distance_cm**2 * Fx  #Luminosity Function

    print(f"\n{name}")
    print(f"Fx = {Fx:.3e} erg cm^-2 s^-1")
    print(f"Lx = {Lx:.3e} erg s^-1")
    print(f"log10(Lx) = {np.log10(Lx):.3f}")
```
