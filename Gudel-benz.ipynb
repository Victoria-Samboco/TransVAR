```python

# ── MKT J2236-2519 ─────────────────────────────────────────────────────
my_distance_pc = 31.193
d_cm = my_distance_pc * pc_to_cm

# ── QUIESCENT: PyBDSF time-averaged flux ─────────────────────────────────────
q_flux_jy     = 1.063321020072e-4 #106.3e-6 #90.17e-6   # Jy
q_flux_err_jy =  2.217360974285e-5 #22.2e-6 #0.202e-3   # Jy RMS stats value
q_radio_lum     = q_flux_jy     * jy_to_erg * 4*np.pi*d_cm**2
q_radio_lum_err = q_flux_err_jy * jy_to_erg * 4*np.pi*d_cm**2

# ── FLARE: peak flux from residual cube lightcurve ───────────────────────────
# Peak flux = 1.2839 mJy from cube residual at pixel X=1750, Y=1298
# Error estimated as RMS noise per timeslot ~150 uJy
f_flux_jy     = 1.332947e-3   # Jy
f_flux_err_jy = 1.37409e-4    # Jy  (per-timeslot RMS)
f_radio_lum     = f_flux_jy     * jy_to_erg * 4*np.pi*d_cm**2
f_radio_lum_err = f_flux_err_jy * jy_to_erg * 4*np.pi*d_cm**2

# ── X-ray: 1RXS J223715.2-262234/ TIC 326446019───────────────────────
my_xray_flux     = 2.06e-12   # erg s-1 cm-2
my_xray_flux_err = 0.0
my_xray_flux_err = 0.2 * my_xray_flux
my_xray_lum      = my_xray_flux     * 4*np.pi*d_cm**2
my_xray_lum_err  = my_xray_flux_err * 4*np.pi*d_cm**2

print(f"--- Quiescent ---")
print(f"L_R  : {q_radio_lum:.3e} erg/s/Hz  (log={np.log10(q_radio_lum):.2f})")
print(f"--- Flare peak ---")
print(f"L_R  : {f_radio_lum:.3e} erg/s/Hz  (log={np.log10(f_radio_lum):.2f})")
print(f"--- X-ray (both) ---")
print(f"L_X  : {my_xray_lum:.3e} erg/s     (log={np.log10(my_xray_lum):.2f})")
```

```python
# ── Colours ───────────────────────────────────────────────────────────────────
pointsize   = 11
grey_pts    = 'Black'
blue_pts    = '#FFFC31'
red_pts     = '#D56AA0'
star_quiet  ="blue" #'#FB8C00' # orange star  — quiescent
star_flare  = "#E31A1C"   #'#E53935'   # red star  — flare peak

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# ── Literature populations ────────────────────────────────────────────────────
ax.errorbar(weighted_mean_erg, dMes_xray,
            xerr=weighted_mean_erg_err, c=red_pts,
            fmt='^', zorder=2)
ax.errorbar(radioerg_1992AA, xrays_1992AA,
            xerr=radioerg_err_1992AA,
            xuplims=radioflux_ul_1992AA, c=blue_pts,
            fmt='d', zorder=2)
ax.errorbar(radioerg_1989ApJS, xray_1989ApJS,
            xerr=radioerg_err_1989ApJS,
            yerr=xray_err_1989ApJS, fmt='o', c=grey_pts,
            xuplims=radio_1989ApJS_ul,
            uplims=xray_1989ApJS_ul, zorder=2)

# ── Güdel-Benz best-fit line ──────────────────────────────────────────────────
test_rs = np.linspace(10**11.5, 10**18., 100)
test_xs = 1.06*np.log10(test_rs) + 14.52
ax.plot(test_rs, 10.**test_xs, 'k--', alpha=0.8, zorder=0)

# ── Quiescent point ───────────────────────────────────────────────────────────
ax.errorbar(q_radio_lum, my_xray_lum,
            xerr=q_radio_lum_err,
            yerr=my_xray_lum_err,
            fmt='*', markersize=24,
            color=star_quiet,
            markeredgecolor='black',
            markeredgewidth=1.2,
            ecolor=star_quiet,
            capsize=4,
            zorder=5)

# ── Flare peak point ──────────────────────────────────────────────────────────
ax.errorbar(f_radio_lum, my_xray_lum,
            xerr=f_radio_lum_err,
            yerr=my_xray_lum_err,
            fmt='*', markersize=24,
            color=star_flare,
            markeredgecolor='black',
            markeredgewidth=1.2,
            ecolor=star_flare,
            capsize=4,
            zorder=5)

# ── Arrow connecting quiescent to flare ──────────────────────────────────────
ax.annotate('',
            xy=(f_radio_lum, my_xray_lum),
            xytext=(q_radio_lum, my_xray_lum),
            arrowprops=dict(arrowstyle='->', color='grey',
                           lw=1.5, linestyle='dashed'))

# ── Labels ────────────────────────────────────────────────────────────────────
ax.annotate('Quiescent',
            xy=(q_radio_lum, my_xray_lum),
            xytext=(q_radio_lum*0.3, my_xray_lum*2.5),
            fontsize=11, color=star_quiet,
            arrowprops=dict(arrowstyle='->', color=star_quiet, lw=1.2))

ax.annotate('Flare peak',
            xy=(f_radio_lum, my_xray_lum),
            xytext=(f_radio_lum*2.5, my_xray_lum*2.5),
            fontsize=11, color=star_flare,
            arrowprops=dict(arrowstyle='->', color=star_flare, lw=1.2))

# ── Legend ────────────────────────────────────────────────────────────────────
dme_dot = mlines.Line2D([], [], markerfacecolor=red_pts, marker='^',
                        markeredgecolor='None', markersize=pointsize+2,
                        label='dM/dMe stars', color='None')
dke_dot = mlines.Line2D([], [], markerfacecolor=blue_pts, marker='d',
                        markeredgecolor='None', markersize=pointsize+2,
                        label='dKe stars', color='None')
rscvn_dot = mlines.Line2D([], [], markerfacecolor=grey_pts, marker='o',
                          markeredgecolor='None', markersize=pointsize+2,
                          label='RS CVn binaries', color='None')
bestfit_line = mlines.Line2D([], [], color='Black', alpha=0.8,
                             linestyle='--',
                             label='Best fit (Güdel et al. 1993)')
quiet_dot = mlines.Line2D([], [], markerfacecolor=star_quiet, linestyle='None', marker='*',
                          markeredgecolor='black', markeredgewidth=1.2,
                          markersize=pointsize+8,
                          label='MKT J2236-2519 (Quiescent)',
                          color='orange')
flare_dot = mlines.Line2D([], [], markerfacecolor=star_flare, linestyle='None', marker='*',
                          markeredgecolor='black', markeredgewidth=1.2,
                          markersize=pointsize+8,
                          label='MKT J2236-2519 (Flare peak)',
                          color='Red')

ax.legend(handles=[rscvn_dot, dme_dot, dke_dot,
                   bestfit_line, quiet_dot, flare_dot],
          fontsize=13, frameon=True, loc='lower right',
          edgecolor='None', facecolor='White')
ax.set_ylabel(r'$L_X\,\,\left[\mathrm{erg\,s^{-1}}\right]$',
              fontsize=22)
ax.set_xlabel(r'$L_R\,\,\left[\mathrm{erg\,s^{-1}\,Hz^{-1}}\right]$',
              fontsize=22)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(10**11.5, 10**18)
ax.set_ylim(10**25.,  10**33.)

fig.savefig('Guedel_Benz_quiescent_flare.png', dpi=600,
            transparent=False, bbox_inches='tight')

fig.savefig('Guedel_Benz_quiescent_flare.pdf', format="pdf", bbox_inches='tight')
plt.show()

print(f"\nQuiescent: log L_R = {np.log10(q_radio_lum):.2f}")
print(f"Flare peak: log L_R = {np.log10(f_radio_lum):.2f}")
print(f"Both:       log L_X = {np.log10(my_xray_lum):.2f}")
```
